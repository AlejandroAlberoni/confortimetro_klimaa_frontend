```python
class CheckPmvConditions:
    def __init__(self, ep_api):
        self.ep_api = ep_api
        
    def __call__(self, state):
        try:
            pmv_handle = self.ep_api.exchange.get_variable_handle(state, "Zone Thermal Comfort Fanger Model PMV")
            if pmv_handle != None:
                print(self.ep_api.exchange.get_variable_value(state, pmv_handle))
        except Exception as ex:
            print("Error")
```

```python
ep_api = EnergyPlusAPI()
state = ep_api.state_manager.new_state()

ep_api.runtime.callback_begin_system_timestep_before_predictor(state, CheckPmvConditions)
ep_api.runtime.run_energyplus(state, ['--weather', EPW_PATH, '--output-directory', OUTPUT_PATH, '-r', IDF_PATH])
ep_api.state_manager.reset_state(state)
```
